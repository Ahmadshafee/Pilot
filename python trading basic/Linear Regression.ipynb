```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model ,datasets 
```

```python
diabetes= datasets.load_diabetes()
print(diabetes.keys())
print(diabetes.data)
print(diabetes.DESCR)
```

```python
diabetes_x= diabetes.data[:,np.newaxis,2]#TO TAKE ONE LABEL OR FEATURE
print(diabetes_x)
```

```python
diabetes_x_train=diabetes_x[-20:]
diabetes_x_test=diabetes_x[:-20]

diabetes_y=diabetes.target
diabetes_y_train=diabetes_x[-20:]
diabetes_y_test=diabetes_x[:-20]
diabetes_y_test

```

```python
regmodel=linear_model.LinearRegression()#to create a linear regression model
regmodel.fit(diabetes_x_train,diabetes_y_train)#to train the model using the training data
```

```python
# Mean squared error
print('Mean squared error: %2f' %
      np.mean((regmodel.predict(diabetes_x_test)-diabetes_y_test)**2))
```

```python
plt.figure(figsize=(10,7))
plt.scatter(diabetes_x_test,diabetes_y_test,color='black')
plt.plot(diabetes_x_test,regmodel.predict(diabetes_x_test),color='blue')
plt.title('Linear Regression Model', fontsize=14)
plt.xlabel('Independent Variable', fontsize=12)
plt.ylabel('Dependent Variable', fontsize=12)
plt.show
```
